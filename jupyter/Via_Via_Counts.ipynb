# Analyze the total amount of exits and entries

```python
File = pd.read_csv(baseDir + "raw/nissan/junction_counts.csv", sep=",")        
df = pd.DataFrame(File)
```

```python
viaEx = pd.read_csv(baseDir + "raw/nissan/enter_count.csv")
viaEn = pd.read_csv(baseDir + "raw/nissan/exit_count.csv")
viaEn = viaEn[viaEn['x'] == viaEn['x']]
viaEx = viaEx[viaEx['x'] == viaEx['x']] 
```

```python
mapp = folium.Map(location=[51.133,10.288],
                  zoom_start = 7
                  )  

Feature_Group_Entry = folium.map.FeatureGroup(name="Entries", overlay=True, control=True)
for i,g in viaEn.iterrows():
    folium.features.Circle(location=[g['y'],g['x']],
                           popup = ("Entry:"+str(g['value'])),
                           radius = g['value']/10,
                           color = "red",
                           fill = True,
                           fill_color = "white"
                          ).add_to(Feature_Group_Entry)
    
Feature_Group_Exit = folium.map.FeatureGroup(name="Exits", overlay=True, control=True)
for i,g in viaEx.iterrows():
    folium.features.Circle(location=[g['y'],g['x']],
                           popup = ("Exit:"+str(g['value'])),
                           radius = g['value']/10,
                           color = "blue",
                           fill = True,
                           fill_color = "white"
                          ).add_to(Feature_Group_Exit)

Feature_Group_Entry.add_to(mapp)
Feature_Group_Exit.add_to(mapp)
folium.LayerControl().add_to(mapp)



mapp
```
